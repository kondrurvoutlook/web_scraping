```python
import itertools
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
products=[]
prices=[]
ratings=[]
num_of_ratings=[]
```

```python
url_list=[]
for i in range(1,41):
  url=f"https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&otracker=categorytree&page={i}"
  url_list.append(url)
#url_list
```

```python
url='https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&otracker=categorytree&page=1'
text=requests.get(url)
text.status_code
```

```python
for url in url_list:
  text=requests.get(url)
  text.status_code
  soup=BeautifulSoup(text.content, "html")
  
  for a in soup.findAll('a',href=True, attrs={'class':'_1fQZEK'}):
    try:
      name=a.find('div', attrs={'class':'_4rR01T'})
      price=a.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
      rating=a.find('div', attrs={'class':'_3LWZlK'})
      numofrating=a.find('span',attrs={'class':'_2_R_DZ'})
      products.append(name.text)
      prices.append(price.text)
      ratings.append(rating.text) 
      num_of_ratings.append(numofrating.text)
    except AttributeError:
      k='0'
      ratings.append(k)
      num_of_ratings.append(k)
      continue
flipkart_data=pd.DataFrame({"ProductName":products,"Price":prices,"Rating":ratings,"Number of Ratings":num_of_ratings})
```

```python
flipkart_data=flipkart_data.loc[flipkart_data['Rating']!='0']
flipkart_data
```
